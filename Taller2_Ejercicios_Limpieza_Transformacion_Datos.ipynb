##Ejercicio Duración del contrato


```python
df_SECOP["Unidad de Duracion"].value_counts()
```

```python
df_SECOP[["Unidad de Duracion","Duracion"]].sample(20)
```

```python
df_SECOP["Duracion_Dias"] = np.where(df_SECOP["Unidad de Duracion"] == "Dias", df_SECOP["Duracion"]/30,  df_SECOP["Duracion"])
df_SECOP["Duracion_Dias"].sample(20)
```

```python
df_SECOP["Duracion_Meses"] = np.where(df_SECOP["Unidad de Duracion"] == "Meses", df_SECOP["Duracion"]/12,  df_SECOP["Duracion"])
df_SECOP["Duracion_Meses"].sample(20)
```

```python
df_SECOP["Duracion_Años"] = np.where(df_SECOP["Unidad de Duracion"] == "Años", df_SECOP["Duracion"]/360,  df_SECOP["Duracion"])
df_SECOP["Duracion_Años"].sample(20)
```
